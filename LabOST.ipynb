```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Student': ['A','B','C','D','E','F','G','H','I','J'],
    'Maths': [85, 90, 78, 88, 76, 95, 89, 92, 70, 80],
    'Science': [82, 91, 74, 85, 79, 94, 88, 90, 72, 84],
    'English': [75, 85, 80, 83, 78, 92, 86, 88, 74, 81],
    'Attendance': [90, 95, 80, 85, 70, 98, 92, 94, 72, 88]
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.describe()
```

```python
# Histogram
df[['Maths','Science','English']].plot(kind='hist', alpha=0.6, bins=8)
plt.title("Histogram of Subject Marks")
plt.show()
```

```python
# Box Plot
sns.boxplot(data=df[['Maths','Science','English']])
plt.title("Box Plot of Marks")
plt.show()
```

```python
# Line Plot
df.set_index('Student')[['Maths','Science','English']].plot(marker='o')
plt.title("Line Plot of Marks by Student")
plt.show()
```

```python
# Bar Plot
df.plot(x='Student', y=['Maths','Science','English'], kind='bar')
plt.title("Bar Plot of Student Marks")
plt.show()
```

```python
# Scatter Plot
plt.scatter(df['Maths'], df['Science'], color='purple')
plt.xlabel("Maths Marks")
plt.ylabel("Science Marks")
plt.title("Scatter Plot: Maths vs Science")
plt.show()
```

```python
# Relplot (Seaborn)
sns.relplot(x='Maths', y='Science', hue='Attendance', size='English', data=df)
plt.title("Relational Plot (relplot)")
plt.show()
```

```python
# Joint Plot
sns.jointplot(x='Maths', y='Science', data=df, kind='reg')
plt.show()
```

```python
# PairGrid (Multivariate plot)
g = sns.PairGrid(df[['Maths','Science','English','Attendance']])
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
plt.show()
```

```python
# Heatmap (Correlation Matrix)
plt.figure(figsize=(6,4))
sns.heatmap(df[['Maths','Science','English','Attendance']].corr(), annot=True, cmap='coolwarm')
plt.title("Heatmap of Correlation")
plt.show()
```
